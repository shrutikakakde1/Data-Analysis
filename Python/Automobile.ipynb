```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
automobile = pd.read_csv('Automobile.csv')
```

```python
automobile.head(10)
```

```python
#Distance plotting with kernal density estimation using continous variable (Univariate Distribution)
#Distplot - Parametric distribution of data
sns.distplot(automobile['wheel_base'])
plt.show()
```

```python
#Distance plotting without kernal density estimation using continous variable (Univariate Distribution) 
#Lines at bottom defines each observation
#Distplot - Parametric distribution of data

sns.distplot(automobile['wheel_base'], kde=False, rug=True)
plt.show()
```

```python
#kind : {scatter | reg | resid | hex | kde}, optional
sns.jointplot(automobile['engine_size'],kind="hex")
plt.show()
```

```python
sns.jointplot(automobile[['wheel_base','engine_size','price']])
```

```python
sns.stripplot(automobile['fuel_type'],jitter=True)
```

```python
sns.swarmplot(automobile['fuel_type'])
```

```python
sns.countplot(automobile['engine_size'])
```

```python
sns.barplot(automobile['horsepower'])
```

```python
sns.pointplot(automobile['horsepower'])
```
